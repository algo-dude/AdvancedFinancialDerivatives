```python
import pandas as pd
import numpy as np

live = pd.read_csv('live.csv')
duke = pd.read_csv('duke.csv')
live.head()
```

```python
# where strategy is not MOO_micro
live = live[live['Strategy'] != 'MOO_micro']

# first day 1/21/2022 with duke
# last day 4/12/2022 with duke
live.DateIn = pd.to_datetime(live.DateIn)
duke.DateIn = pd.to_datetime(duke.DateIn)

# # remove all live trades before 1/21/2022
live = live[live.DateIn >= ('2022-01-21')]


live.head()

```

```python
print(len(live), 'live trades excluding other systems not tradable with paper limitations')
print(len(duke), 'paper trades')

print(round(live.PctPL.sum(), 4),' % PnL live excluding other systems not tradable with paper limitations')
print(round(duke.PctPL.sum(), 4),' % PnL paper ')
# note: not compunded
```
