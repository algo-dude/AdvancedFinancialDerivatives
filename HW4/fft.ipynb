```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)

def my_function(x):
    return -(1/3)*(x**3) + (1/2)*(x**2) - (3/16)*x
```

```python
# find 9th order fourier approximation of my_function
def fourier_approx(x, n):
    sum = 0
    for k in range(n):
        sum += my_function(x) * (2*np.pi*k/n) * np.cos(2*np.pi*k*x/n)
    return sum
```

```python
interval = np.linspace(start=0, stop=1, num=100)
fn = my_function(interval)
fax = fourier_approx(interval, 9)
plt.plot(fn, label='my_function')
plt.plot(fax, label='fourier_approx')
plt.legend()
plt.show()
```
